```python
import networkx as nx
from scipy.stats import pearsonr
```

Questão 1: Alternativa E

```python
#Lê o arquivo
with open("USairport500.txt", 'r') as file:
  arestas=[]
  for linha in file:
    vertices=linha.strip().split()
    arestas.append((vertices[0],vertices[1]))

#Cria o grafo usando o arquivo
G=nx.Graph(arestas)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
G = nx.convert_node_labels_to_integers(G, first_label=0)
G=G.subgraph(max(nx.connected_components(G),key=len))

#Calcula o eigenvector e exibe o resultado
eigenvector=list(nx.eigenvector_centrality(G).values())
print(sum(eigenvector)/len(eigenvector))
```

Questão 2: Alternativa E

```python
#Função que recebe um grafo e retorna um vetor com o grau de cada nó
def obterGraus(G):
  aux=[]
  for node in G.nodes():
    aux.append(len(list(G.neighbors(node))))
  return aux;

#Lê o arquivo
with open("hamsterster.txt", 'r') as file:
  arestas=[]
  for linha in file:
    vertices=linha.strip().split()
    arestas.append((vertices[0],vertices[1]))

#Cria o grafo usando o arquivo
G=nx.Graph(arestas)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
G = nx.convert_node_labels_to_integers(G, first_label=0)
G=G.subgraph(max(nx.connected_components(G),key=len))

#Exibe o resultado
print(pearsonr(list(nx.betweenness_centrality(G).values()),obterGraus(G)))
```

Questão 3: Alternativa C

```python
from scipy.stats import spearmanr

#Lê o arquivo
with open("jazz.txt", 'r') as file:
  arestas=[]
  for linha in file:
    vertices=linha.strip().split()
    arestas.append((vertices[0],vertices[1]))

#Cria o grafo usando o arquivo
G=nx.Graph(arestas)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
G = nx.convert_node_labels_to_integers(G, first_label=0)
G=G.subgraph(max(nx.connected_components(G),key=len))

#Calcula e exibe o resultado
print(spearmanr(list(nx.closeness_centrality(G).values()),list(nx.core_number(G).values())))
```

-É calculada a partir da exponencial da matriz de probabilidade de transição: Random Walk

-Considera os menores caminhos entre os pares de vértices: Closeness Centrality

-É calculada a partir do espectro da matriz de adjacência: Eigenvector Centrality

-Usa os menores caminhos de modo a determinar a carga( em termos de quantidade de tráfego) em cada vértice: Betweenness Centrality

-É uma medida de centralidade local, pois alguns vértices com valores altos dessa medida podem estar na borda da rede: Degree Centrality

-Um vértice é considerado central se estiver conectado a outros centrais. É usada em redes dirigidas: Page Rank
