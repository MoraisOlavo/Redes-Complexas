```python
import networkx as nx
from networkx.algorithms.community import greedy_modularity_communities, modularity, louvain_communities
```

Questão 1: Alternativa B

```python
#Lê o arquivo e cria o grafo
G=nx.read_edgelist("jazz.txt",nodetype=int)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
G = nx.convert_node_labels_to_integers(G, first_label=0)
G=G.subgraph(max(nx.connected_components(G),key=len))

#Calcula a modularidade usando Greedy como algoritmo para detectar comunidades
print(modularity(G,greedy_modularity_communities(G)))
```

Questão 2: Alternativa E

```python
#Lê o arquivo e cria o grafo
G=nx.read_edgelist("jazz.txt",nodetype=int)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
G = nx.convert_node_labels_to_integers(G, first_label=0)
G=G.subgraph(max(nx.connected_components(G),key=len))

#Calcula a modularidade usando Louvain como algoritmo para detectar comunidades
print(modularity(G,louvain_communities(G)))
```

Questão 3: Alternativa A

```python
#Função que cria o grafo de acordo com os parâmetros passados e retorna a modularidade usando louvain como algoritmo de detecção de comunidades
def GerarGrafoCalcularModularidade(N,tau1,tau2,mu,k,minc,maxc):
  G=nx.LFR_benchmark_graph(n=N,tau1=tau1,tau2=tau2,mu=mu,min_degree=k,max_degree=k,min_community=minc,max_community=maxc,seed=10)
  print("u= "+str(mu)+", Q(u)= "+str(modularity(G,louvain_communities(G))))

#Cria os grafos e exibe os resultados
GerarGrafoCalcularModularidade(128,3,1.5,0.05,16,32,32)
GerarGrafoCalcularModularidade(128,3,1.5,0.1,16,32,32)
GerarGrafoCalcularModularidade(128,3,1.5,0.2,16,32,32)
```
