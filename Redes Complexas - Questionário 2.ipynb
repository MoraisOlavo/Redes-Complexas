Questão 1: Alternativa C

```python
import networkx as nx
#lê o arquivo e armazena as arestas em um array arestas

#Função que recebe um Grafo e retorna as menores distâncias entre todos os pares de nós
def calcularDistancias(G):
  dist=[]
  for i in range(0,G.number_of_nodes()):
    for j in range(i+1,G.number_of_nodes()):
      dist.append(len(nx.shortest_path(G, source=i, target=j))-1)

  return dist

#Lê o arquivo
with open("hamsterster.txt", 'r') as file:
  arestas=[]
  for linha in file:
    vertices=linha.strip().split()
    arestas.append((vertices[0],vertices[1]))

#cria o grafo a partir do arquivo
G=nx.Graph()
G=nx.Graph(arestas)
G=G.subgraph(max(nx.connected_components(G),key=len))
G = nx.convert_node_labels_to_integers(G, first_label=0)

#Exibe os resultado
dist=calcularDistancias(G)
print("Diâmetro da rede: "+str(max(dist)))
print("Médias dos menores caminhos: "+str(sum(dist)/len(dist)))
```

```python
import math

#Função que recebe um vetor e calcular a entropia de Shannon dos elementos do vetor
def entropiaShannon(v):
  #Encontra o maior elemento de v
  maior=0
  for i in v:
    if i>maior:
      maior=i

  #Conta quantas vezes cada elemento aparece
  contador=[]
  for i in range(0,maior+1):
    contador.append(0)

  for i in v:
    contador[i]+=1

  #Calcula a entropia de Shannon
  soma=0
  for i in range(0,maior+1):
    if(contador[i]!=0):
      soma+=contador[i]/len(v)*math.log2(contador[i]/len(v))

  return -1*soma

#Lê o arquivo
with open("USairport500.txt", 'r') as file:
  arestas=[]
  for linha in file:
    vertices=linha.strip().split()
    arestas.append((vertices[0],vertices[1]))

#Cria o grafo a partir do arquivo
G=nx.Graph()
G=nx.Graph(arestas)
G=G.subgraph(max(nx.connected_components(G),key=len))
G = nx.convert_node_labels_to_integers(G, first_label=0)

#Calcula e exibe o resultado
dist=calcularDistancias(G)
print("Entropia de Shannon: "+str(entropiaShannon(dist)))
```
